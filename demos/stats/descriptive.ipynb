```python
import pandas as pd
import numpy as np
import json
import gzip
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'axes.axisbelow': True})
```

# Toy data

```python
num_of_likes = pd.Series([5, 8, 10, 12, 15, 100, 200])
```

```python
plt.hist(num_likes);
plt.xlabel("Number of likes")
plt.ylabel("Frequency")
```

```python
print(num_likes.mean())
print(num_of_likes.median())
```

# Real data

```python
with gzip.open("botwiki-2019_tweets.json.gz") as f:
    botwiki = json.load(f)
```

```python
botwiki_df = pd.DataFrame.from_records([item['user'] for item in botwiki])
```

```python
plt.hist(botwiki_df.followers_count)
plt.xlabel("Number of followers")
plt.ylabel("Frequency")
```

```python
np.logspace(1, 6.2, num=20)
```

```python
plt.hist(
    botwiki_df.followers_count,
    bins=np.logspace(1, 6.2, num=20)
);
plt.xlabel("Number of followers")
plt.ylabel("Frequency")
```

```python
plt.hist(
    botwiki_df.followers_count,
    bins=np.logspace(1, 6.2, num=20)
);
plt.xscale('log')
plt.xlabel("Number of followers")
plt.ylabel("Frequency")
```

```python
plt.hist(
    botwiki_df.followers_count,
    bins=np.logspace(1, 6.2, num=20)
);
plt.axvline(botwiki_df.followers_count.mean(), color="black")
plt.axvline(botwiki_df.followers_count.median(), color="red")

plt.xscale('log')
plt.xlabel("Number of followers")
plt.ylabel("Frequency")
```

```python
print(botwiki_df.followers_count.mean())
print(botwiki_df.followers_count.median())
```

```python
plt.ecdf(
    botwiki_df.followers_count,
);
plt.xlabel("Number of followers")
plt.ylabel("CDF")
```

```python
plt.ecdf(
    botwiki_df.followers_count,
);
plt.xscale('log')
plt.xlabel("Number of followers")
plt.ylabel("CDF")
```

```python
plt.ecdf(
    botwiki_df.followers_count,
    complementary=True
);
plt.xscale('log')
plt.xlabel("Number of followers")
plt.ylabel("CCDF")
```
