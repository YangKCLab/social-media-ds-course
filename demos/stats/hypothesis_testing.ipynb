```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'axes.axisbelow': True})
```

```python
import scipy.stats
```

```python
def simulate_coin_toss(bias=0.5, n_trials=1000):
    """
    Simulate coin toss outcomes with a specified bias.

    Parameters:
    -----------
    bias : float
        Probability of getting a head (1). Must be between 0 and 1.
    n_trials : int
        Number of tosses to simulate.

    Returns:
    --------
    dict
        A dictionary with:
        - 'frequency': {'head': count, 'tail': count}
        - 'average': average value (proportion of heads)
    """
    if not (0 <= bias <= 1):
        raise ValueError("Bias must be between 0 and 1.")
    if n_trials <= 0:
        raise ValueError("Number of trials must be positive.")

    # Generate toss outcomes: 1 for head, 0 for tail
    outcomes = [1 if random.random() < bias else 0 for _ in range(n_trials)]
    
    # Compute frequency and average
    heads = sum(outcomes)
    tails = n_trials - heads
    avg = heads / n_trials

    return {
        'frequency': {'head': heads, 'tail': tails},
        'average': avg
    }
```

```python
outcome = simulate_coin_toss(0.5, 100_000)
plt.bar(
    [0, 1],
    [
        outcome['frequency']['tail'],
        outcome['frequency']['head']
    ],
    
),
plt.xticks([0, 1], ["Tail (0)", "Head (1)"])
plt.ylabel("Frequency")
```

# Repeat the experiment

```python
fair_coin_results = []
biased_coin_results = []
for i in range(100_000):
    fair_result = simulate_coin_toss(0.500, 100)
    biased_result = simulate_coin_toss(0.600, 100)
    fair_coin_results.append(fair_result['average'])
    biased_coin_results.append(biased_result['average'])
```

```python
plt.hist(fair_coin_results, bins=40, histtype='step');
plt.hist(biased_coin_results, bins=40, histtype='step');
plt.xlabel("Average value of the toss")
plt.ylabel("Frequency")
```

```python
scipy.stats.ttest_ind(fair_coin_results, biased_coin_results)
```
