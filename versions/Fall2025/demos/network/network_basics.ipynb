```python
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'axes.axisbelow': True})
```

# Creating and drawing a undirected graph

```python
# A "plain" graph is undirected
G = nx.Graph()

# give each node a 'name', which is a letter in this case.
G.add_node('a')

# The add_nodes_from method allows adding nodes from a sequence, in this case, a list
nodes_to_add = ['b', 'c', 'd']
G.add_nodes_from(nodes_to_add)

# add edge from 'a' to 'b'
# since this graph is undirected, the order doesn't matter here
G.add_edge('a', 'b')

# just like add_nodes_from, we can add edges from a sequence
# edges should be specified as 2-tuples
edges_to_add = [('a', 'c'), ('b', 'c'), ('c', 'd')]
G.add_edges_from(edges_to_add)

# draw the graph
nx.draw(G, with_labels=True)
```

# Graph methods

```python
# List all the nodes
G.nodes()
```

```python
# List all the edges
G.edges()
```

```python
# Number of nodes
```

```python
G.number_of_nodes()
```

```python
# Number of edges
G.number_of_edges()
```

```python
# Find the neighbors of a node "b"
G.neighbors('b')
```

```python
for node in G.neighbors("b"):
    print(node)
```

# Node degree

```python
# You can simply count the number of neighbors of a node
len(list(G.neighbors("a")))
```

```python
# Or, you can use the function G.degree
G.degree("a")
```

```python
# You can generate the degree distribution of this network
degree_sequence = [d for _, d in G.degree()]
degree_sequence = pd.Series(degree_sequence)
```

```python
degree_sequence
```

```python
plt.hist(degree_sequence)
plt.xlabel("Degree")
plt.ylabel("Frequency")
```
